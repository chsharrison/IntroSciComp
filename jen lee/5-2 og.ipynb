```python
# instantiating defualts
plt.figure(figsize=(8, 6), dpi=60)
plt.subplot(1, 1, 1)
x = np.linspace(-np.pi, np.pi, 256, endpoint=True)
c, s = np.cos(x), np.sin(x)
plt.plot(x, c, color='blue', linewidth=2.5, linestyle="-", label="cosine")
plt.plot(x, s, color='green', linewidth=2.5, linestyle="-", label="sine")
plt.legend(loc='upper left')

# adding annotations to points
t = 2 * np.pi / 3
plt.plot([t, t], [0, np.cos(t)], color='blue', linewidth=2.5, linestyle="--")
plt.scatter([t, ], [np.cos(t), ], 50, color='blue')

plt.annotate(r'$cos(\frac{2\pi}{3})=-\frac{1}{2}$',
             xy=(t, np.cos(t)), xycoords='data',
             xytext=(-90, -50), textcoords='offset points', fontsize=16,
             arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))

plt.plot([t, t],[0, np.sin(t)], color='green', linewidth=2.5, linestyle="--")
plt.scatter([t, ],[np.sin(t), ], 50, color='red')

plt.annotate(r'$sin(\frac{2\pi}{3})=\frac{\sqrt{3}}{2}$',
             xy=(t, np.sin(t)), xycoords='data',
             xytext=(+10, +30), textcoords='offset points', fontsize=16,
             arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))

# setting ticks and axis limits
plt.xlim(x.min(), x.max())
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi], [r'$-\pi$', r'$-\pi/2$', r'$0$', 
                                                  r'$+\pi/2$', r'$\pi/2$', r'$\pi$'])
plt.ylim(c.min() * 1.1, c.max() * 1.1)
plt.yticks([-1, 0, 1])
plt.show()
```

# 1.4.4.3 Bar Plots


```python
n = 12 
x = np.arange(n)
y1 = (1 - x / float(n)) * np.random.uniform(0.5, 1.0, n)
y2 = (1 - x / float(n)) * np.random.uniform(0.5, 1.0, n)
plt.bar(x, y1, facecolor='#9999ff', edgecolor='k')
plt.bar(x, -y2, facecolor='#ff9999', edgecolor='k')
for x, y in zip(x, y1):
    plt.text(x + 0.4, y + 0.05, '%.2f' % y, ha='center', va='bottom')
plt.ylim(-1.25, 1.25)
```

# 1.4.4.4 Contour Plots

```python
def f(x, y):
    return (1 - x / 2 + x ** 5 + y ** 3) * np.exp(-x ** 2 - y ** 2) 

n = 256
x = np.linspace(-3, 3, n)
y = np.linspace(-3, 3, n)
x, y = np.meshgrid(x, y)

plt.axes([0.025, 0.025, 0.95, 0.95])

plt.contourf(x, y, f(x, y), 8, alpha=.75, 
             cmap=plt.cm.hot)
c = plt.contour(x, y, f(x, y), 8, colors='black', linewidth=.5)

plt.clabel(c, inline=1, fontsize=10)
plt.xticks(())
plt.yticks(())
plt.show()
```

# 1.4.4.7 Quiver Plots

```python
n = 8 
k, y, = np.mgrid[0:n, 0:n]
t = np.arctan2(y - n / 2., k - n/2.)
r = 10 + np.sqrt((y - n / 2.) ** 2 + (k - n / 2.) ** 2)
u, v = r * np.cos(t), r * np.sin(t)
plt.axes([0.025, 0.025, 0.95, 0.95])
plt.quiver(k, y, u, v, r, alpha=.7) # colors inside arrows
plt.quiver(k, y, u, v, edgecolor='r', facecolor='None', linewidth=.5)# colors outside arrows
plt.xticks(())
plt.yticks(())
plt.ylim(-1, n)
plt.xlim(-1, n)
```

# 1.4.4.10 Polar Axis

```python
ax = plt.axes([0, 0, 1, 1], polar=True)
n = 20
theta = np.arange(0, 2 * np.pi, 2 * np.pi / n)
radii = 10 * np.random.rand(n)
width = np.pi / 4 * np.random.rand(n)
bars = plt.bar(theta, radii, width=width, bottom=0)
for r, bar in zip(radii, bars):
    bar.set_facecolor(plt.cm.cool(r / 10))
    bar.set_alpha(0.5)
ax.set_yticklabels(())
ax.set_xticklabels(())
```

# 1.4.4.11 3D Plots

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = Axes3D(fig)
x = np.arange(-4, 4, 0.25)
y = np.arange(-4, 4, 0.25)
x, y = np.meshgrid(x, y)
r = np.sqrt(x**2 + y**2)
z = np.sin(r)
ax.plot_surface(x, y, z, rstride=1, cstride=1, cmap='cool')
ax.contourf(x, y, z, zdir='z', offset=-2, cmap=plt.cm.cool)
ax.set_zlim(-2, 2)
```
