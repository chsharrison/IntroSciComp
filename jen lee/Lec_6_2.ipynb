```python
import numpy as np

import matplotlib.pyplot as plt
```

```python
a=np.zeros((16,20))
a
np.full(((16,20)),3)
```

```python
a[:]=3
```

```python
a[0:2,0:2]==6
```

```python
a[0:2,0:2]==6
```

```python

plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.pcolormesh(a,cmap='cool')
plt.xlabel('x-axis')
plt.ylabel('y-label')
plt.title('Mesh Title')
cbar=plt.colorbar()
cbar.set_label('label')

plt.subplot(1,2,2)
plt.imshow(a,cmap='cool') # the axis is diffrent in each plot
plt.xlabel('x-axis')
plt.ylabel('y-label')
cbar.set_label('colorbar')
plt.title('Imshow Title')
cbar=plt.colorbar()
```

```python
# The x-axis in colormesh represents a larger data set, the graph in imshow doesnt show the number 16 or 20
# the y- axis in imshow represents the 2x2 square ploted diffrently
```

```python
methods=[None,'none','nearest','bilinear','bicubic','spline16','spline36','hanning','hamming','hermite','kaiser','quadric','catrom', 'gaussian', 'bessel', 'mitchell', 'sinc', 'lanczos']


np.random.seed(19680801)
grid=np.random.rand(4,4)

fig,axs=plt.subplots(nrows=4,ncols=4, figsize(9,6),
                     subplot_kw={'xticks':[],'yticks':[]})
for ax, interp_method in zip (axs.flat, methods, cmap='cool')
    axs set_title(str(interp_method))
plt.tight_layout()
plt.show
```
