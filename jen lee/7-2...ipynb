### 3.1 
From the movies spreadsheet you used last week (Lec 6.2), make a histogram of all the years movies were made, combining the data in three sheets. In one plot, make two subplots, one with bins (bars) of 10 years width, and one with bins of 20 years width. Note this does not mean 10 and 20 bins!


```python
#import the file
excel_file = 'movies.xls'
movies = pd.read_excel(excel_file)#pandas reads file
# deal with the separate sheets
movies_sheet1 = pd.read_excel(excel_file, sheet_name = 0, index_col = 0)
movies_sheet1.head()# fisrt spreadsheet
movies_sheet2 = pd.read_excel(excel_file, sheetname=1, index_col=0)
movies_sheet2.head()#second
movies_sheet3 = pd.read_excel(excel_file, sheetname=2, index_col=0)
movies_sheet3.head()#third
movies = pd.concat([movies_sheet1, movies_sheet2, movies_sheet3])#all together now
#combine the sheets into one df
xlsx = pd.ExcelFile(excel_file)
movies_sheets = []
for sheet in xlsx.sheet_names:
    movies_sheets.append(xlsx.parse(sheet))
    movies = pd.concat(movies_sheets)
```

```python
movies.Year.plot(kind = 'hist', bins = 50)
```

```python
import numpy as np
```

```python
bins10 = np.arange(1916,2020,10)
bins20 = np.arange(1916,2020,20)

plt.subplot(2, 1, 1)
movies.Year.plot(kind = 'hist', bins=bins10,color='blue', edgecolor='green')#outlines can be changed colors just like the color of the bars
plt.xlabel('10 Bins')
plt.ylabel('10  Bins')

plt.subplot(2, 1, 2)
movies.Year.plot(kind = 'hist', bins = bins20,color='red', edgecolor='green')
plt.ylabel('20 Bins')
```
