## 1.3.1.6 Copies and views

```python
a = np.arange(10)
a
```

```python
b = a[::2]
b
```

```python
np.may_share_memory(a, b)
```

```python
b[0] = 12
b
```

```python
a   # note that changing b[0] also changed a[0]
```

```python
a = np.arange(10)
```

```python
c = a[::2].copy()  # force a copy
c[0] = 12
c
```

```python
a #notice that forcing a copy does not change a[0]
```

```python
np.may_share_memory(a, c)
#check to see if they share memory
```

```python
is_prime = np.ones((100,), dtype=bool)
```

```python
is_prime[:2] = 0
```

```python
N_max = int(np.sqrt(len(is_prime) - 1))
for j in range(2, N_max + 1):
    is_prime[2*j::j] = False
```
