## Example: 
    Plot archimedian spiral  

R code for this spiral

a <- 1.5

b <- -2.4

t <- seq(0,5*pi, length.out=500)

x <- (a + b*t) * cos(t)

y <- (a + b*t) * sin(t)

plot(x,y, type="l", col=2, lwd=3)

abline(h=0, v=0, col="grey")

```python
a = 1.5
b = -2.4
t = np.linspace(0,5*np.pi, 500)
x = (a + b*t)*np.cos(t)
y = (a + b*t)*np.sin(t)
plt.plot(x,y)
plt.axis('equal') #makes plot square
plt.grid() #gives a grid 
```

```python
a = 1.5
b = 2.4 #change from negative to positive
t = np.linspace(0,5*np.pi, 500)
x = (a + b*t)*np.cos(t)
y = (a + b*t)*np.sin(t)
plt.plot(x,y)
plt.axis('equal') 
plt.grid() 
```
