```python
#make a 3x5 arra of random mubers
x = np.random.rand(3, 5)
x
```

```python
#plot the array - what range are the values in? range = 0.0-1.0
plt.pcolormesh(x)
plt.colorbar()
```

```python
#what is the standard dev for each axis


#the standard deviation of axis 0 gives the standard deviation of each column within the original array (x)
#standard deviation describes how much the individual values within the group vary from the group mean 
#the size of the resulting array is 1x5 
#the first and third columns have the most variation within them as thier sd is higher
##column number five has the smallest sd indicating that all the values in this column were close to one another


sd_col = x.std(axis = 0) 
sd_col
```

```python
plt.plot(sd_col)
```

```python
#the standard deviation of axis 1 gives the standard deviation of each row within the original array (x) 
#the size of the resulting array is 1x3 
#the second row has the most variation as it has the highest sd
##row number one has the least variation

sd_row = x.std(axis = 1) 
sd_row
```

```python
plt.plot(sd_row)
```

```python
#what is the mean for each axis

#the mean of axis 0 gives the average of all the values in each column within the original array (x) 
#the size of the resulting array is 1x5 
#the second and third columns had the lowest values on average
#the fifth column had the highest values


mean_col = x.mean(axis = 0)
mean_col
```

```python
plt.plot(mean_col, 'k', linewidth = 2)
# plot +/- std
plt.plot(mean_col+sd_col, 'k--')
plt.plot(mean_col - sd_col, 'k--')
plt.xlabel('Column')
plt.ylabel('Mean +/- std')
plt.grid()
```

```python
#the mean of axis 1 gives the average of all the values in each row within the original array (x) 
#the size of the resulting array is 1x3 
#the Frist row had the lowest values on average
#the third row had the highest values
#in this plot we see a general increasing trend, if this was real data and not random values this might be interesting

mean_row = x.mean(axis = 1)
mean_row
```

```python
plt.plot(mean_row)
```

## Masking

Masking allows you to exclude data without getting rid of it
sub-package of numpy called np.ma


```python
x2 = np.ma.masked_where(x < 0.5, x) #the x after the comma is the data that is being masked
```

```python
plt.pcolormesh(x2, vmin = 0, vmax = 2)
plt.colorbar()
#the white areas are those that are masked so they are being ignored
```

```python
x2.sum() #ignores the masked values 
```

```python
x.sum() #masked values are included
```

```python
plt.hist(x2)
```

```python
# now mask the y array and re-plot the histogram 
y2 = np.ma.masked_where(y < 0.5, y)
```

```python
plt.pcolormesh(y2, vmin = 0, vmax = 2)
plt.colorbar()
```

```python
#scatter plot
n = 1024
X = np.random.normal(0, 1, n)
Y = np.random.normal(0, 1, n)
T = np.arctan2(Y, X)

plt.axes([0.025, 0.025, 0.95, 0.95])
plt.scatter(X, Y, s=75, c=T, alpha=.5)

plt.xlim(-1.5, 1.5)
plt.xticks(())
plt.ylim(-1.5, 1.5)
plt.yticks(())

plt.show()
```

```python
#bar plot 
n = 12
X = np.arange(n)
Y1 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)
Y2 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)

plt.axes([0.025, 0.025, 0.95, 0.95])
plt.bar(X, +Y1, facecolor='#9999ff', edgecolor='white')
plt.bar(X, -Y2, facecolor='#ff9999', edgecolor='white')

for x, y in zip(X, Y1):
    plt.text(x + 0.4, y + 0.05, '%.2f' % y, ha='center', va= 'bottom')

for x, y in zip(X, Y2):
    plt.text(x + 0.4, -y - 0.05, '%.2f' % y, ha='center', va= 'top')

plt.xlim(-.5, n)
plt.xticks(())
plt.ylim(-1.25, 1.25)
plt.yticks(())

plt.show()
```

```python
#contour plot
def f(x,y):
    return (1 - x / 2 + x**5 + y**3) * np.exp(-x**2 -y**2)

n = 256
x = np.linspace(-3, 3, n)
y = np.linspace(-3, 3, n)
X,Y = np.meshgrid(x, y)

plt.axes([0.025, 0.025, 0.95, 0.95])

plt.contourf(X, Y, f(X, Y), 8, alpha=.75, cmap=plt.cm.hot)
C = plt.contour(X, Y, f(X, Y), 8, colors='black', linewidth=.5)
plt.clabel(C, inline=1, fontsize=10)

plt.xticks(())
plt.yticks(())
plt.show()
```

```python
#imshow
def f(x, y):
    return (1 - x / 2 + x ** 5 + y ** 3 ) * np.exp(-x ** 2 - y ** 2)

n = 10
x = np.linspace(-3, 3, 3.5 * n)
y = np.linspace(-3, 3, 3.0 * n)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

plt.axes([0.025, 0.025, 0.95, 0.95])
plt.imshow(Z, interpolation='nearest', cmap='bone', origin='lower')
plt.colorbar(shrink=.92)

plt.xticks(())
plt.yticks(())
plt.show()
```

```python
#Quiver plot
n = 8
n = 8
X, Y = np.mgrid[0:n, 0:n]
T = np.arctan2(Y - n / 2., X - n/2.)
R = 10 + np.sqrt((Y - n / 2.0) ** 2 + (X - n / 2.0) ** 2)
U, V = R * np.cos(T), R * np.sin(T)

plt.axes([0.025, 0.025, 0.95, 0.95])
plt.quiver(X, Y, U, V, R, alpha=.5)
plt.quiver(X, Y, U, V, edgecolor='k', facecolor='None', linewidth=.5)

plt.xlim(-1, n)
plt.xticks(())
plt.ylim(-1, n)
plt.yticks(())

plt.show()
```
