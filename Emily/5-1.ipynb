## Go over 1.3.5.3

```python
data = np.loadtxt('populations.txt')
year, hares, lynxes, carrots = data.T
populations = data[:,1:] #getting rid of the first column ofyear data
```

```python

print("       Hares, Lynxes, Carrots")
print("Mean:", populations.mean(axis=0))
print("Std:", populations.std(axis=0))
```

```python
#argmax returns the idicies of the maximun values along an as
j_max_years = np.argmax(populations, axis=0)
print("Max. year:", year[j_max_years])
```

```python
np.argmax(populations, axis=0)
#hares max at index 3, linx at 4, carrots at 0
#check with plot, this looks right
```

```python
#sp with largest pop each year 
```

```python
populations
```

```python
#get index for the population that is largest each year
np.argmax(populations, axis = 1)
```

```python
#make an array of names
max_species = np.argmax(populations, axis=1)
species = np.array(['Hare', 'Lynx', 'Carrot'])
print("Max species:")
print(year)
print(species[max_species])
```

```python
above_50000 = np.any(populations > 50000, axis=1)
print("Any above 50000:", year[above_50000])
```

```python
j_top_2 = np.argsort(populations, axis=0)[:2]
print("Top 2 years with lowest populations for each:")
print(year[j_top_2])
```

```python
hare_grad = np.gradient(hares, 1.0)
print("diff(Hares) vs. Lynxes correlation", np.corrcoef(hare_grad, lynxes)[0,1])
```

```python

import matplotlib.pyplot as plt
plt.plot(year, hare_grad, year, -lynxes)
plt.savefig('plot.png')
```

## Exercise 1

Plot a logarithmic spiral

```python
import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt
from math import exp,sin,cos
from pylab import *

mpl.rcParams['legend.fontsize'] = 10

fig = plt.figure()
ax = fig.gca(projection='3d')
n=100
a=0.5
b=0.20
th=np.linspace(0, 500, 10000)
x=a*exp(b*th)*cos(th)
y=a*exp(b*th)*sin(th)
ax.plot(x, y)
ax.legend()

plt.show()
```
