### W 2

W 2.1 Plot the height function $H = -100(lat-40)^2 -400{(lon+106)^2} + 8000$ over 38 to 42 N and 108 to 104 W. 

W 2.2 Label your axes, add a colorbar, and make the fonts readable for a presentation (aka larger). 

W 2.3 What mountain range is this a rough approximation to? Add this to the title.

W 2.4 (Advanced) plot this in 3D

```python
# make one dimensinal arrays using linspace
lat = np.linspace(38,42,100) 
lon = np.linspace(-108,-104,100)
```

```python
plt.plot(lat)
```

```python
plt.plot(lat)
```

```python
#use meshgrid to to combine the two arrays
X,Y = np.meshgrid(lon,lat)
```

```python
#add height function 
H = (-100*(Y-40)**2) - 400*(X+106)**2 + 8000
```

```python
plt.pcolormesh(X,Y,H)

plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Approximation of the Rocky Mountains')
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.plot_surface(X,Y,H, cmap = 'winter')
```
