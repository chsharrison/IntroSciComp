W 2.1 Plot the height function $H = -100(lat-40)^2 -400{(lon+106)^2} + 8000$ over 38 to 42 N and 108 to 104 W. 

W 2.2 Label your axes, add a colorbar, and make the fonts readable for a presentation (aka larger). 

W 2.3 What mountain range is this a rough approximation to? Add this to the title.

```python
lat = np.linspace(38,42,100) #these are 1-D arrays
lon = np.linspace(-108,-104,100)


# H = -100((lat-40)**2) - 400((lon+106)**2) + 8000
```

```python
plt.plot(lat)
```

```python
plt.plot(lon)
```

```python
X,Y = np.meshgrid(lon,lat)
```

```python
H = (-100*(Y-40)**2) - 400*(X+106)**2 + 8000
```

```python
plt.pcolormesh(X,Y,H)

mpl.rcParams['font.size'] = 16
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Rocky Mountains')
```

```python
import pandas as pd
```

```python
excel_file = 'movies.xls'
movies = pd.read_excel(excel_file)
```

```python
movies.head()
```

```python
#to call a page
movies_sheet1 = pd.read_excel(excel_file, sheetname=0, index_col=0)
movies_sheet1.head()
```

```python
movies_sheet2 = pd.read_excel(excel_file, sheetname=1, index_col=0)
movies_sheet2.head()
```

```python
movies_sheet3 = pd.read_excel(excel_file, sheetname=2, index_col=0)
movies_sheet3.head()
```

```python
#to combine all the movies together use concat
movies = pd.concat([movies_sheet1, movies_sheet2, movies_sheet3])
```

```python
movies.shape
```

```python
#anothr option is to wrap the file then pass it
##more convenient if you have a lot of sheets
xlsx = pd.ExcelFile(excel_file)
movies_sheets = []
for sheet in xlsx.sheet_names:
    movies_sheets.append(xlsx.parse(sheet))
    movies = pd.concat(movies_sheets)
```
