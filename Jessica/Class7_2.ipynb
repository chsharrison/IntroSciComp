### 3.2
Label your axes and change the color of the bars in each plot. Make the bars have black outlines.

```python
plt.subplot(1,2,1)
movies.Year.plot(kind = 'hist', bins=10, color = 'green', edgecolor = 'black', linewidth = 1) 
plt.xticks(ten_year)
plt.xlabel('10 Year Count')
plt.ylabel('Number of Movies')

plt.subplot(1,2,2)
movies.Year.plot(kind = 'hist',bins=5, color = 'purple', edgecolor = 'black', linewidth = 2) 
plt.xticks(twenty_year)
plt.xlabel('20 Year Count')
plt.ylabel('Number of Movies')

plt.subplots_adjust(bottom=0.1, right=2, top=0.9)
```
