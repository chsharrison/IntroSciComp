### Exercise: 
    1) plot sin(x) on -pi to pi using 100 points
    2) plot sin(x*y) on a grid using 100x100 points on (-pi,pi) * (-pi,pi)

```python
plt.figure(figsize=(8,6), dpi=80)

x = np.linspace(-np.pi, np.pi, 100, endpoint=True)
s = np.sin(x)

plt.plot(x,s, color='blue', linewidth = 1.0, linestyle='-')

plt.xlim(x.min() * 1.1, x.max() * 1.1)
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
          [r'$-\pi$',r'$-\pi/2$',r'$0$',r'$\pi/2$',r'$\pi$'])


plt.yticks([-1,0,+1])
```

```python
x.shape
```

```python
plt.figure(figsize=(8,6), dpi=80)

x = np.linspace(-np.pi, np.pi, 100, endpoint=True)
y = np.linspace(-np.pi, np.pi, 100, endpoint=True)
X,Y = np.meshgrid(x,y)
s = np.sin(X*Y)

plt.pcolormesh(X,Y,s)


plt.xlim(x.min() * 1.1, x.max() * 1.1)
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
          [r'$-\pi$',r'$-\pi/2$',r'$0$',r'$\pi/2$',r'$\pi$'])

plt.ylim(y.min() * 1.1, y.max() * 1.1)
plt.yticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
          [r'$-\pi$',r'$-\pi/2$',r'$0$',r'$\pi/2$',r'$\pi$'])

plt.colorbar()
```

```python
plt.figure(figsize=(8,6), dpi=80)

x = np.linspace(-np.pi, np.pi, 100, endpoint=True)
y = np.linspace(-np.pi, np.pi, 100, endpoint=True)
X,Y = np.meshgrid(x,y)
s = np.sin(X*(Y**2))

plt.pcolormesh(X,Y,s)


plt.xlim(x.min() * 1.1, x.max() * 1.1)
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
          [r'$-\pi$',r'$-\pi/2$',r'$0$',r'$\pi/2$',r'$\pi$'])

plt.ylim(y.min() * 1.1, y.max() * 1.1)
plt.yticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
          [r'$-\pi$',r'$-\pi/2$',r'$0$',r'$\pi/2$',r'$\pi$'])

plt.colorbar()
```

# Archimedean spiral example

```python
a = 1.5
b = -2.4
t = np.linspace(0,5*np.pi,500)
x = (a + b*t) * np.cos(t)
y = (a + b*t) * np.sin(t)
```

```python
plt.plot(x,y)
plt.axis('equal') #makes the plot square so tick marks are equal
#plt.plot([-40,40], [0,0], color = 'grey')
plt.grid()
```
