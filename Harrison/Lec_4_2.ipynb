```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.random.rand(3,5)
```

```python
x
```

```python
plt.pcolormesh(x)
plt.colorbar()
```

```python
# this gives you the mean of each column (5 means)
col_mean = np.mean(x, axis = 0)
```

```python
# standard deviation of each column
col_std = np.std(x,axis = 0)
```

```python
plt.plot(col_mean, 'k', linewidth = 2)
plt.plot(col_mean + col_std, 'k--')
plt.plot(col_mean - col_std, 'k--')
plt.xlabel('column')
plt.ylabel('mean +/- standard deviation')
plt.title('Column statistics', fontsize = 12)
plt.grid()
```

Similar method for the rows

```python
# this gives you the mean of each row (3 means)
np.mean(x, axis = 1)
```

etc

```python
# Using logical indexing, add one to all the values greater than 0.5
x[x>0.5]+= 1
```

```python
plt.pcolormesh(x)
plt.colorbar()
```

```python
# this gives you the mean of each column (5 means)
col_mean = np.mean(x, axis = 0)
# standard deviation of each column
col_std = np.std(x,axis = 0)
plt.plot(col_mean, 'k', linewidth = 2)
plt.plot(col_mean + col_std, 'k--')
plt.plot(col_mean - col_std, 'k--')
plt.xlabel('Column')
plt.ylabel('Mean +/- standard deviation')
plt.title('Column statistics', fontsize = 12)
plt.grid()
```

```python
#Plot the 2X2 bottom left hand corner of the array

plt.pcolormesh(x[0:2,0:2])
plt.colorbar()
```

```python
y = np.random.rand(300,500)
```

```python
plt.pcolormesh(y)
plt.colorbar()
```

```python
# let's use a histogram to look at the data:
# you have to flatten the array to make a vector, otherwise it will try to make a histogram of each row
plt.hist(y.flatten())
```

```python
# add one to all of the values above 0.5:
y[y>0.5] += 1
# or y[y>0.5] = y[y>0.5]+1
```

```python
plt.hist(y.flatten())
```
