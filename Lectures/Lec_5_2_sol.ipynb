## 2. Warm-up:

Code a function that calculates the volume of a cube, given the length of a side. Use this to plot how volume scales with length of a side. Label your plot.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# trick to change the plotting settings:
import matplotlib as mpl
mpl.rcParams['font.size'] = 16
mpl.rcParams['lines.linewidth']= 2.0
```

```python
# volume is V = s^3

def vol(s):
    return s**3
    
```

```python
# check:
side = 30
vol(side)
```

```python
#plot:

plt.figure(figsize=(5,5))

side = np.arange(0,100)
plt.plot(side, vol(side))

plt.xlabel('Side length')
plt.ylabel('Volume')
plt.title('Volume of a cube')
```
