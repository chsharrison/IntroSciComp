### 6. Compare (plot) the change in hare population (see help(np.gradient)) and the number of lynxes. Check correlation (see help(np.corrcoef))

```python
hares
```

```python
np.gradient(hares)
# this is a rate, population change per year
```

```python
plt.plot(np.gradient(hares), label = 'hares grad')
plt.plot(lynxes, label = 'lynxes')
plt.legend(loc = 'upper center')
# these appear to be anti-correlated, when lynxes go up, the hares gradient goes down (i.e. decreases)
```

```python
plt.scatter(np.gradient(hares),lynxes)
# that's a pretty good correlation
plt.xlabel('Hares gradient (hares/year)')
plt.ylabel('Lynxes')
plt.title('r = -0.92')
```

```python
np.corrcoef(np.gradient(hares), lynxes)
# the number we want is the -0.92
```

## Exercise 2

Plot the hares/carrots/lynxes data as three scatterplots as subplots in one plot, comparing hares and lynxes, hares and carrots, and carrots and lynxes. Do a (Pearson) correlation analysis for each plot/data pares (google correlation). Display your correlation coefficient on each graph.

```python
plt.figure(figsize = (4,20))
plt.subplot(3,1,1)
plt.scatter(hares,carrots)
plt.xlabel('Hares')
plt.ylabel('Carrots')
plt.title('r = %.3f' %np.corrcoef(hares, carrots)[0,1])
# see https://www.learnpython.org/en/String_Formatting

plt.subplot(3,1,2)
plt.scatter(hares,lynxes)
plt.xlabel('Hares')
plt.ylabel('Lynxes')
plt.title('r = %.3f' %np.corrcoef(hares, lynxes)[0,1])

plt.subplot(3,1,3)
plt.scatter(carrots,lynxes)
plt.xlabel('Carrots')
plt.ylabel('Lynxes')
plt.title('r = %.3f' %np.corrcoef(carrots, lynxes)[0,1]);
```
