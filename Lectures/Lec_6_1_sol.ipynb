Plot an Archimedean spiral (https://en.wikipedia.org/wiki/Archimedean_spiral)

R code:

a <- 1.5

b <- -2.4

t <- seq(0,5*pi, length.out=500)

x <- (a + b*t) * cos(t)

y <- (a + b*t) * sin(t)

plot(x,y, type="l", col=2, lwd=3)

abline(h=0, v=0, col="grey")

```python
# python translation:
a = 1.5
b = -2.4
t = np.linspace(0,5*np.pi,500)
x = (a+b*t)*np.cos(t)
y = (a+b*t)*np.sin(t)
plt.plot(x,y)
#plt.plot([-40, 40], [0,0], color = 'grey')
plt.axis('equal') # makes the plot square, i.e. tick marks are equally spaced
# https://matplotlib.org/3.1.1/gallery/subplots_axes_and_figures/axis_equal_demo.html
plt.grid()
plt.title('Archemedian spiral')
;
```

```python
# messing around with the parameters
a = 1.5
b = 2.4
t = np.linspace(0,5*np.pi,500)
x = (a+b*t)*np.cos(t)
y = (a+b*t)*np.sin(t)
plt.plot(x,y)
plt.plot([-40, 40], [0,0], color = 'grey')
plt.axis('equal') # makes the plot square, i.e. tick marks are equally spaced
# https://matplotlib.org/3.1.1/gallery/subplots_axes_and_figures/axis_equal_demo.html
plt.grid()
```

## Exercise 1 (Saved from Lec 5.1)

Plot a logarithmic spiral (google it)

Equation is:

x = a*e^(kt)*cos(t)

y = a*e^(kt)*sin(t)


Using the same technique as above:

```python
a = 1
k = 0.17
t = np.linspace(0,10*np.pi,1000)
x = a*np.exp(k*t)*np.cos(t)
y = a*np.exp(k*t)*np.sin(t)
plt.plot(x,y)
plt.axis('equal')
plt.title('Logarithmic spiral')
;
```
