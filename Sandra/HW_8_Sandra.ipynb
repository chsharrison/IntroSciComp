```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import seaborn as sns
```

```python
from sklearn import datasets ## imports datasets from scikit-learn
```

```python
data = datasets.load_boston()
```

```python
# convert the data set into a pandas data frame  
df = pd.DataFrame(data.data, columns=data.feature_names)

# Put the target (housing value -- MEDV) in another DataFrame
target = pd.DataFrame(data.target, columns=["MEDV"])

df["MEDV"] = target["MEDV"] 
```

```python
from scipy import stats
slope, intercept, r_value, p_value, std_err = stats.linregress(df["AGE"],target["MEDV"]) #linear regression #scipy
```

```python

plt.plot(df["AGE"], slope*df["AGE"]+intercept, 'k--')
plt.scatter(df["AGE"], target["MEDV"])
plt.xlabel('Age of Home')
plt.ylabel('House value ($1000s)')
plt.title('Home Value vs Age')
plt.rcParams.update({'font.size': 14})
```
