```python
def fred (a):
    if a >= 0:
        r = np.sqrt(a/3.14)
        cir = 2 * 3.14 * r
        print("radius:", r, "circumference:", cir)
    else: 
        print("error this area is negative")
```

Warm up 1.3

```python
area = np.arange(0, 1001,20)
def fred (a):
    r = np.sqrt(a/3.14)
    cir = 2 * 3.14 * r
    plt.plot(a,r, label = "radius")
    plt.plot(a, cir, label = "circumference")
    plt.title('radius and circumference of a circle')
    plt.xlabel('area of a circle')
    plt.ylabel('radius and circumference length')
    plt.legend()
```

```python
fred(area)
```
