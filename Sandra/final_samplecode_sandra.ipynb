```python
import pandas as pd
```

```python
from matplotlib import pyplot as plt
```

```python
data = pd.read_excel(r'Bag\RedYearBag.xlsx')
```

```python
data2 = pd.read_excel(r'Bay\RedYearBay.xlsx')
```

```python
data3 = pd.read_excel(r'Gill\RedYearGill.xlsx')
```

```python
df = pd.DataFrame(data, columns= ['station','Number of Fish Caught','Latitude','Longitude'])
```

```python
df2 = pd.DataFrame(data2, columns= ['station','Number of Fish Caught','Latitude','Longitude'])
```

```python
df3 = pd.DataFrame(data3, columns= ['station','Number of Fish Caught','Latitude','Longitude'])
```

```python
fig, ax = plt.subplots()
df.plot(kind="scatter", x="Longitude", y="Latitude", marker='s', c="Number of Fish Caught", cmap="viridis", ax=ax);
plt.title("2008-2018 Bag Seine Red Drum")
```

```python
fig2, ax = plt.subplots()
df2.plot(kind="scatter", x="Longitude", y="Latitude", marker='s', c="Number of Fish Caught", cmap="viridis", ax=ax);
plt.title("2008-2018 Bay Trawl Red Drum")
```

```python
fig3, ax = plt.subplots()
df3.plot(kind="scatter", x="Longitude", y="Latitude", marker='s', c="Number of Fish Caught", cmap="viridis", ax=ax);
plt.title("2008-2018 Gill Net Red Drum")
```
