## Exercise 1 (Saved from Lec 5.1)

Plot an archimedean spiral (google it)

```python
import numpy as np
import matplotlib.pyplot as plt
a = 1.5
b = 2.4
t = np.linspace(0,5*np.pi,500)
x = (a+b*t)*np.cos(t)
y = (a+b*t)*np.sin(t)
plt.plot(x,y)
plt.plot([-40, 40], [0,0], color = 'grey')
plt.axis('equal')
plt.grid()
```

```python
a = 1
k = 0.17
t = np.linspace(0,10*np.pi,1000)
x = a*np.exp(k*t)*np.cos(t)
y = a*np.exp(k*t)*np.sin(t)
plt.plot(x,y)
plt.axis('equal')
plt.title('Logarithmic spiral')
;
```

x= (vt+c)cos(wt)
y= (vt+c)sin(wt)

where v, c, w are all constants and t is the parameter that varies to make the spiral
