```python
# 1 The mean and std of the populations of each 
# species for the years in the period.
import matplotlib.pyplot as plt
data = np.loadtxt('populations.txt')
year, hares, lynxes, carrots = data.T # Transpose, turns rows to column
populations = data[:, 1:]
pop_mean = populations.mean(axis=0)
pop_std = populations.std(axis=0)
```

```python
# 2 Which year each species had the largest population.
pop_max_year = np.argmax(populations,axis=0)
year[pop_max_year]
```

```python
# 3 Which species has the largest population for each year. 
# (Hint: argsort & fancy indexing of np.array(['H', 'L', 'C']))
pop_max = np.argmax(populations, axis=1)
species = np.array(['hares', 'lynxes', 'carrots'])
pop_max
species[pop_max]
```

```python
# 4 Which years any of the populations is above 50000. 
# (Hint: comparisons and np.any)
above_5000 = np.any(populations > 50000, axis=1)
year[above_5000]
```

```python
# 5 The top 2 years for each species when they had the lowest populations. 
# (Hint: argsort, fancy indexing)
pop_min = np.argsort(populations, axis=0) [:2]
year[pop_min]
```

```python
# Compare (plot) the change in hare population (see help(np.gradient)) 
# and the number of lynxes. Check correlation (see help(np.corrcoef)).

hare_grad = np.gradient(hares, 1.0)
print("diff(Hares) vs. Lynxes correlation", np.corrcoef(hare_grad, lynxes)[0,1])

import matplotlib.pyplot as plt
plt.plot(year, hare_grad, year, lynxes)
plt.savefig('plot.png')
```
